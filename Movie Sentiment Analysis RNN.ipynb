```
import nltk

nltk.download('movie_reviews')
```

```
from nltk.corpus import movie_reviews

neg = movie_reviews.fileids('neg')
```

```
negative_fileids = movie_reviews.fileids('neg')
positive_fileids = movie_reviews.fileids('pos')
```

```
negative_fileids
```

```
#reading the entire corpus into this and creating labels as we go..
text =[]
labels = []
for i in range(1000):
  text.append(movie_reviews.raw(fileids = positive_fileids[i]))
  labels.append(0)
  
for i in range(1000):
  text.append(movie_reviews.raw(fileids = negative_fileids[i]))
  labels.append(1)
```

```
print(text[2], labels[2])
```

```
import numpy as np
text = np.array(text)
labels = np.array(labels)
```

```
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

max_len = 600 #this is for the padding lengths
training = int(2000*.8)
testing = int(2000* .2)

max_words = 15000
#first we create a tokenizer
tokenizer = Tokenizer(num_words = max_words)
#then we fit it onto the text
tokenizer.fit_on_texts(text)
#finally we can change those texts to integer
sequences = tokenizer.texts_to_sequences(text)

# now we pad the sequences so that they are all the same shape
data = pad_sequences(sequences, maxlen = max_len)

labels = np.asarray(labels)

#now we split the data into training and testing 
#remember that the data is organized into [0-1000] = pos, [1000-2000] = neg

indices = np.arange(data.shape[0]) #this takes the form of the indices
# now we split the data

np.random.shuffle(indices)
data = data[indices]
labels = labels[indices]

X_train = data[:training]
X_test= data[training:]

y_train = labels[:training]
y_test = labels[training:]

from keras.utils import to_categorical

y_train1 = to_categorical(y_train)
y_test1 = to_categorical(y_test)



```

```
from keras import Model
from keras.models import Sequential
from keras import layers
from keras import regularizers


model = Sequential()
model.add(layers.Embedding(max_words, 128))
# model.add(layers.LSTM(128, activation = 'relu', recurrent_dropout = .2, return_sequences=True, kernel_regularizer = regularizers.l2(0.001)))
# model.add(layers.Dropout(.2))
# model.add(layers.LSTM(128, activation = 'relu', recurrent_dropout = .2, return_sequences=True, kernel_regularizer = regularizers.l2(0.001)))
# model.add(layers.Dropout(.1))
# model.add(layers.LSTM(64, activation = 'relu', recurrent_dropout = .2, return_sequences=True,kernel_regularizer = regularizers.l2(0.001)))
# model.add(layers.Dropout(.2))
# model.add(layers.LSTM(64, activation = 'relu', recurrent_dropout = .2, return_sequences=True,kernel_regularizer = regularizers.l2(0.001)))
# model.add(layers.LSTM(32, activation = 'relu', recurrent_dropout = .2))
model.add(layers.GRU(256, return_sequences = True, recurrent_dropout = .2, kernel_regularizer = regularizers.l2(0.001)))
model.add(layers.Dropout(.2))
# model.add(layers.GRU(128, return_sequences = True, recurrent_dropout = .2, kernel_regularizer = regularizers.l2(0.001)))
# model.add(layers.Dropout(.4))
model.add(layers.GRU(64, return_sequences = True, recurrent_dropout = .4, kernel_regularizer = regularizers.l2(0.001)))
model.add(layers.Dropout(.2))
model.add(layers.GRU(32))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics = ['acc'])
history = model.fit(X_train, y_train,
                   epochs = 10,
                   batch_size = 128,
                    validation_split = .7
                  )
```
