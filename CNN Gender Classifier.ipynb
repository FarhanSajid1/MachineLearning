```
from keras import models
from keras import layers
from keras import optimizers
import os
from keras.preprocessing.image import ImageDataGenerator
from keras.applications import VGG16
import numpy as np
```

```
import zipfile
import io
data = zipfile.ZipFile(io.BytesIO(uploaded['gendertesting.zip']), 'r')
data.extractall()

train_dir = 'gendertesting/train/'   
test_dir = 'gendertesting/test/' 
data.printdir()
```

```
from keras import models
from keras import layers
from keras import optimizers

conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(150, 150, 3))

model = models.Sequential()
model.add(conv_base) #we are adding the base model from before
model.add(layers.Flatten()) #we are flattening manually
model.add(layers.Dense(256, activation = 'relu'))
model.add(layers.Dense(2, activation = 'softmax'))

model.summary()

train_datagen = ImageDataGenerator(
  rescale = 1./255,
    rotation_range = 60,
    width_shift_range = .2,
    height_shift_range = .2,
    shear_range = .2,
    zoom_range = .2,
    horizontal_flip = True,
    fill_mode = 'nearest'
)

test_datagen = ImageDataGenerator(rescale = 1./255)

train_generator= train_datagen.flow_from_directory(
  train_dir,
    target_size = (150,150),
    batch_size = 40,
    classes = ['male', 'female']
)

validation_generator = test_datagen.flow_from_directory(
  test_dir,
    target_size = (150,150),
    batch_size = 25,
    classes = ['male', 'female']
)

model.compile(loss = 'categorical_crossentropy',
             optimizer=optimizers.RMSprop(lr=2e-5),
             metrics = ['acc'])

history = model.fit_generator(
  train_generator, 
    steps_per_epoch = 100 ,
    epochs = 40,
    validation_data = validation_generator,
    validation_steps = 50

)


model.save('male_or_female.h5')
```
